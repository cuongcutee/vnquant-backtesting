```python
from vnstock import Vnstock


def get_data(universe):
    data = {}
    for symbol in universe:
        stock = Vnstock().stock(symbol= symbol , source='KBS')
        df = stock.quote.history(start='2020-01-01', end='2026-04-17')
        data[symbol] = df
    return data


def get_datafield(data):
    data_new = {}
    first_symbol = list(data.keys())[0]
    datafield_list = data[first_symbol].columns

    for datafield in datafield_list:
    # Tạo dict {mã: Series} cho mỗi loại dữ liệu
        datafield_dict = {}
        for symbol, df in data.items():
            datafield_dict[symbol] = df[datafield]

        # Chuyển dict thành DataFrame
        data_new[datafield] = pd.DataFrame(datafield_dict)
    
    return data_new


        
        
```

```python
from vnstock import Vnstock
import pandas as pd

stock = Vnstock().stock(symbol='VCI', source='KBS')
df = stock.quote.history(start='2018-01-01', end='2026-04-23')
trading_days = pd.DatetimeIndex(df.index)  # đây là VN trading calendar

# T+2: offset 2 ngày giao dịch thực tế
def add_trading_days(date, n, trading_days):
    idx = trading_days.searchsorted(date)
    return trading_days[idx + n]
```
